## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#keeping age between 18-100
data=data[(data['Age']>=18) & (data['Age']<=100)]
````

````python
data['Gender'].value_counts()
````

````python
#Standardising Gender Value
def clean_gender(gender):
    gender=str(gender).strip().lower()
    if gender in ['male','m', 'man', 'cis male', 'malr', 'mail', 'make']:
        return 'Male'
    elif gender in ['female', 'f', 'woman', 'cis female', 'femake']:
        return 'Female'
    else :
        return 'other'

data['Gender']=data['Gender'].apply(clean_gender)
        
    
````

````python
data['Gender'].value_counts()
````

````python
# Dropping State column becuase it only have US states,for other countries it showing NAN
data.drop(columns=['state'],inplace=True)
````

````python
data['self_employed'].fillna('No',inplace=True)
data.fillna({'work_interfere': "Don't know"}, inplace=True)
````

````python
data.isnull().sum()
````

````python
#Checking and sorting binary and multicategorical columns
categorical_cols=data.select_dtypes(include=object).columns
````

````python
unique_count=data[categorical_cols].nunique().sort_values()
unique_count
````
